**Q1. What is the difference between a function and a method in Python?**

    -> In Python, the main difference between a function and a method is that a function is independent and not tied to any object, while a method is associated with an object (usually defined inside a class). Both are used to perform specific tasks, but methods operate on the data within an object. For example, a simple function can be defined as def add(a, b): return a + b, and you call it using add(2, 3). In contrast, a method is defined inside a class like this:

    class Calculator:
    def add(self, a, b):
        return a + b

**Q5. What are iterators in Python and how do they differ from iterables?**

    -> an iterable is any object that can be looped over (like lists, tuples, strings), while an iterator is an object that produces values one at a time from an iterable, using the next() function.

    An iterable has the __iter__() method, which returns an iterator. An iterator has both __iter__() and __next__() methods. The main difference is that iterables can be re-used (you can loop over them multiple times), but iterators are consumed as you loop through them.

     nums = [1, 2, 3]      
     it = iter(nums)        
     print(next(it))        
     print(next(it))     


**Q6.  Explain the concept of generators in Python and how they are defined.**

    ->  a generator is a special type of function that returns an iterator and allows you to iterate over a sequence of values lazily, meaning it generates values one at a time as needed, instead of storing them all in memory. Generators are defined like regular functions but use the yield keyword instead of return.

    def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
     
     When you call count_up_to(3), it doesn't run the function immediately. Instead, it returns a generator object. You can then iterate through it using a loop or next():

     gen = count_up_to(3)
     print(next(gen))  
     print(next(gen))


**Q7. What are the advantages of using generators over regular functions?**

    -> Generators offer several advantages over regular functions, especially when dealing with large data or streams:

   Memory Efficiency: Generators yield one item at a time, so they don't store the entire result in memory. This is ideal for large datasets or infinite sequences.

   Lazy Evaluation: Generators compute values only when needed, which makes them faster and more responsive in many scenarios.

   Simpler Code: Generators can replace complex iterator classes with simpler syntax using yield.

   Maintain State Automatically: Generators automatically save their state between yields, so you don’t need to manage loop counters or variables manually.

   Examples,

   def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

**Q9. Explain the purpose and usage of the `map()` function in Python**

    -> The map() function in Python is used to apply a function to each item in an iterable (like a list or tuple) and return a new iterator with the results. It helps simplify code by avoiding explicit loops when transforming data.

      Syntax:
         map(function, iterable)

      Example:
          def square(x):
            return x * x

         numbers = [1, 2, 3, 4]
         squared = map(square, numbers)

         print(list(squared))


**Q10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

    -> map(), reduce(), and filter() are built-in functions used for processing collections, but each serves a different purpose:

    map()
      Purpose: Applies a function to each item in an iterable.

      Returns: A new iterator with transformed items.

      Example:
         nums = [1, 2, 3]
         result = map(lambda x: x * 2, nums)
         print(list(result))  


    filter()
       Purpose: Filters items in an iterable based on a condition (function that returns True or False).

       Returns: Only the items for which the function returns True.

       Example:  
         nums = [1, 2, 3, 4]
         result = filter(lambda x: x % 2 == 0, nums)
         print(list(result))


    reduce() (from functools module)
        Purpose: Repeatedly applies a function to reduce the iterable to a single value.

        Returns: A single cumulative result.

        Example:     
            from functools import reduce
           nums = [1, 2, 3, 4]
           result = reduce(lambda x, y: x + y, nums)
           print(result)

**Q1.  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.**

```python
def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum
my_list = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(my_list)
print("Sum of even numbers:", result)
```

**Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self  # Returns the iterator object itself

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration  # Ends the iteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.count += 1
            self.a, self.b = self.b, self.a + self.b
            return self.b


fib = FibonacciIterator(10)
for num in fib:
    print(num, end=" ")
```

**Q6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def powers_of_two(max_exp):
    for i in range(max_exp + 1):
        yield 2 ** i
for power in powers_of_two(5):
    print(power, end=" ")
```

**Q7. Implement a generator function that reads a file line by line and yields each line as a string.**

```python
def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()

with open('example.txt', 'w') as f:
    f.write("Hello\nWorld\nPython\n")


for line in read_file_line_by_line('example.txt'):
    print(line)

```

**Q9.  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.**

```python
celsius = [0, 20, 37, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print("Fahrenheit:", fahrenheit)
```

**Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.**

```python
def remove_vowels(text):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, text))

input_string = "Hello World"
result = remove_vowels(input_string)
print("String without vowels:", result)
```
